# Regression Analysis

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import pandas as pd

# Select relevant features and handle missing values
selected_features = ['length_ft', 'year', 'class']
model_data = data[selected_features + ['price_usd']].dropna()

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    model_data[selected_features],
    model_data['price_usd'],
    test_size=0.2,
    random_state=42
)

# Combine training and test sets for one-hot encoding
combined_data = pd.concat([X_train, X_test])

# One-hot encode 'class' for combined data
combined_data_encoded = pd.get_dummies(combined_data, columns=['class'], drop_first=True)

# Split back into training and test sets
X_train_encoded = combined_data_encoded.iloc[:len(X_train)]
X_test_encoded = combined_data_encoded.iloc[len(X_train):]

# Train a linear regression model
model = LinearRegression()

# Continue with model fitting
model.fit(X_train_encoded, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test_encoded)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
```

### Explanation

Mean Squared Error is a metric used to measure the average squared difference between the actual values (y_test) and the predicted values (y_pred). A lower MSE indicates better predictive accuracy, as it reflects smaller errors between the predicted and actual values. The result of the evaluation will be printed as part of the output, providing insight into how well the model performs on the test dataset.



```python
# Print the range of actual boat prices
print(f'Actual Price Range: {data["price_usd"].min()} - {data["price_usd"].max()} USD')

# Visualize predicted vs. actual prices
plt.scatter(y_test, y_pred)
plt.xlabel('Actual Prices (USD)')
plt.ylabel('Predicted Prices (USD)')
plt.title('Actual vs. Predicted Boat Prices')
plt.show()
```

```python
# Scale numerical features
scaler = StandardScaler()
X_train_encoded_scaled = scaler.fit_transform(X_train_encoded)
X_test_encoded_scaled = scaler.transform(X_test_encoded)

# Continue with model fitting
model.fit(X_train_encoded_scaled, y_train)

# Make predictions on the scaled test set
y_pred_scaled = model.predict(X_test_encoded_scaled)

# Evaluate the model with scaled data
print(f'Mean Squared Error (Scaled): {mean_squared_error(y_test, y_pred_scaled)}')

# Visualize predicted vs. actual prices with scaled data
plt.scatter(y_test, y_pred_scaled)
plt.xlabel('Actual Prices (USD)')
plt.ylabel('Predicted Prices (USD)')
plt.title('Actual vs. Predicted Boat Prices (Scaled Data)')
plt.show()
```

```python
# Feature Importance
feature_importance = pd.Series(model.coef_, index=X_train_encoded.columns)
print('Feature Importance:')
print(feature_importance)
```

### Explanation

The provided output illustrates the feature importance values for each feature in the linear regression model. Feature importance is denoted by coefficients, signifying the extent to which each feature contributes to predicting the target variable (in this instance, boat price in USD). Positive coefficients imply a positive correlation with the target variable, whereas negative coefficients imply a negative correlation.

1. length_ft: This feature has a positive importance of approximately 16826.49, indicating that as the length of the boat increases, the price tends to increase.
2. year: This feature has a positive importance of approximately 2364.39, suggesting that newer boats tend to have higher prices.
3. Various class features: Each class feature represents a type of boat. Positive coefficients indicate a positive correlation with boat price, while negative coefficients indicate a negative correlation.
